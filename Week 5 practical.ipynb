## Exercise
The time rate of change of a population $P(t)$ with constant birth and death rate is, in many simple cases, proportional to the size of the population. That is,

$$ \frac{dP}{dt} = kP $$

where $k$ is the constant of the net growth rate. 

Given the initial population size, $P(0) = 100$, and the net growth rate, $k = 0.1$, 

1) write a program to solve the equation numerically (using Euler's method) to obtain the approximated solution.

2) Plot figures to compare the numerical results to the analytical (exact) results over a time period of 20, considering three different step sizes in Euler's method: 2, 0.5, and 0.1 (Below is an example of the figure with the step size 0.5). 

3) Observe your figures and consider whether the numerical method is accurate and how the accuracy changes with the step size.

Question 1:

```python
# Program to solve the equation (dP/dt = kP), where P(0)=100 and k=0.1

# Define the parameters
# k: net growth rate
# P_init: initial population
# step_size: the time step size
# time: the given time period
# steps: the number of steps during the time period


def euler_method(k, P_init, step_size, time):
    """General equation for using Euler's method:
    P(t+1) = P(t) + step_size * (dP/dt)"""

    steps = int(time/step_size) # number of steps
    P = np.zeros(steps) # create an array to store the population
    P[0] = P_init # set the initial population
    for i in range(1, steps): # loop through the steps
        P[i] = P[i-1] + step_size * k * P[i-1] # calculate the population at the next step
    return P # return the population array

def plot_growth(P, step_size, time):
    # Plot the population growth
    plt.plot(np.arange(0, time, step_size), P) # plot the population
    plt.xlabel('Time')
    plt.ylabel('Population')
    plt.title('Population Growth')
    plt.show()

# Use euler_method to plot
#EXAMPLE1
P_init, k, step_size, time = 100, 0.1, 2, 20
P = euler_method(k, P_init, step_size, time)
plot_growth(P, step_size, time)

#EXAMPLE2
P_init, k, step_size, time = 100, 0.1, 0.5, 20
P = euler_method(k, P_init, step_size, time)
plot_growth(P, step_size, time)

#EXAMPLE3
P_init, k, step_size, time = 100, 0.1, 0.1, 20
P = euler_method(k, P_init, step_size, time)
plot_growth(P, step_size, time)
```
